```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python

# read data
air = pd.read_csv("air_quality.csv" , sep = ',' )
air
```

```python

air['date'] = pd.to_datetime(air['date'])
# get data from 2013 to 2015
air= air [ (air ['date'] > '2002 - 12 - 31' ) &  (air ['date'] < '2016 - 1 - 1') ]
air
```

```python
# retrieve only the columns that matters in our study
air = air[['date', 'so2', 'no2' , 'rspm' ,'spm']].copy()
```

```python
air['Year'] = air['date'].dt.year

air = air.groupby('Year')['so2' ,'no2' ,'rspm' , 'spm'].mean()

```

```python
temperature= pd.read_csv("temperatures.csv" , sep = ',' )
temperature
```

```python
temperature = temperature [(temperature['YEAR'] > 2002)  ]
```

```python
air['Year'] = air.index
fig, axes= plt.subplots(2 , 2 , figsize=(20, 12)  )
  
air_so2 = air[['so2' , 'Year']].dropna()
air_no2 = air[['no2' , 'Year']].dropna()
air_rspm = air[['rspm' , 'Year']].dropna()


sns.barplot(y="so2", x="Year", palette='coolwarm', data=air_so2, ax=axes[0,0] )
axes[0,0].set_title("Average annual so2 in India from 2003 to 2015")

sns.barplot(y="no2", x="Year",palette='rocket', data=air_no2, ax=axes[0,1] )
axes[0,1].set_title("Average annual no2 in India from 2003 to 2015")

sns.barplot(y="rspm", x="Year", palette='mako',data=air_rspm, ax=axes[1,0] )
axes[1,0].set_title("Average annual rspm in India from 2003 to 2015 ")

sns.barplot(y="ANNUAL", x="YEAR", palette = 'mako', data=temperature, ax=axes[1,1])
axes[1,1].set_title("Average annual temperatures in India from 2003 to 2015 ")
plt.savefig ('chart.png')


plt.tight_layout()
               

```
